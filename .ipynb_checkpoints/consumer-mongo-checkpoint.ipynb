```python
from kafka import KafkaConsumer
from pymongo import MongoClient
import json

# Setup kafka consumer
consumer = KafkaConsumer(
    'transaction',
    bootstrap_servers=['localhost:9092'],
    value_deserializer = lambda x: json.loads(x.decode('utf-8'))
)

# setup MongoDB Connection
client = MongoClient("mongodb://admin:password@localhost:27017/")
db = client['fraud_detection']
collection = db['transactions']

# Konsumsi pesan dari kafka dan simpan ke MongoDB
for message in consumer:
    transaction = message.value
    print(f"Menerima pesan: {transaction}")

    # Menghapus field _id jika ada
    if '_id' in transaction:
        del transaction['_id']

    # Definisikan filter untuk mencari dokumen yang sudah ada
    filter = {'transaction_id': transaction.get('transaction_id')}  # Ganti dengan field unik Anda

    try:
        # Update dokumen yang ada, atau insert jika tidak ada
        result = collection.update_one(filter, {'$set': transaction}, upsert=True)
        print("Data telah disimpan ke MongoDB")
    except Exception as e:
        print(f"Error menyimpan data ke MongoDB: {e}")

# Setelah proses penyimpanan ke MongoDB, tambahkan log seperti ini:
if isinstance(data.to_dict('index')[0], list):
    collection.insert_many(data.to_dict('index')[0])
    print("Data batch berhasil disimpan ke MongoDB.")
else:
    collection.insert_one(data.to_dict('index')[0])
    print("Data berhasil disimpan ke MongoDB:", data.to_dict('index')[0])
```
